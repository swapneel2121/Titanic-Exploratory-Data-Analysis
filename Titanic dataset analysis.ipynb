### Data Cleaning

### NULL VALUES

```python
df.isnull().sum()
```

```python
df=df.drop('Cabin',axis=1)
```

```python
df.isnull().sum()
```

```python
df['Age'].skew()
```

```python
df["Age"]=df['Age'].fillna(df['Age'].mean())
```

```python
df.isnull().sum()
```

```python
df['Embarked']=df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df.isnull().sum()
```

### Bining

```python
bins=[0,18,36,72]
label=['child','adult','old']
df['Agegroup']=pd.cut(df['Age'],bins=bins,labels=label)
```

```python
df.head()
```

```python
c=pd.DataFrame(df.groupby('Agegroup')['Survived'].mean()*100)
c
```

### Data Visualization

```python
sns.barplot(x='Sex',y='Survived',data=a,color='red',edgecolor='green')
plt.title('No of males and females Survived')
plt.show()
```

```python
sns.barplot(x="Pclass",y="Survived",data=b,color="orange",edgecolor="black")
plt.title(" Pclass wise survival rate ")
plt.xlabel("Pclass")
plt.ylabel(" Survived Rate")
plt.show()
```

```python
sns.barplot(x="Agegroup",y="Survived",data=c,color="yellow",edgecolor="black")
plt.title(" Agegroup wise survival rate ")
plt.xlabel("Agegroup")
plt.ylabel(" Survived Rate")
plt.show()
```

### Dashboard Creation

```python
plt.figure(figsize=(20,10))
plt.title("Dashboard")
plt.subplot(2,3,1)
sns.countplot(x="Survived",data=df)
plt.subplot(2,3,2)
sns.countplot(x="Sex",hue="Survived",data=df)
plt.subplot(2,3,3)
sns.countplot(x="Pclass",hue='Survived',data=df)
plt.subplot(2,3,4)
sns.countplot(x='Agegroup',hue='Survived',data=df)
plt.subplot(2,3,5)
sns.scatterplot(x='Age',y='Fare',hue='Survived',data=df)
plt.subplot(2,3,6)
sns.boxplot(x='Fare',data=df)
plt.show()
```
